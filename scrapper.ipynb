```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'https://www.imdb.com/search/title/?count=100&groups=top_1000&sort=user_rating'
```

```python
titles = []
ranking = []
years = []
durations = []
votes_numbers = []
scores = []
metascores = []
gross = []
```

```python
def extract_movies(movies, titles, ranking, years, durations, votes_numbers, scores,
                  metascores, gross):
    for movie in movies:
        header = movie.find('h3', attrs={'class': 'lister-item-header'})
        year = header.find('span', attrs={'lister-item-year text-muted unbold'})
        years.append(year)
        titles.append(header.a.text)
        ranking.append(header.span.text)

        time = movie.find('span', attrs={'class' : 'runtime'})
        durations.append(time.text)

        score = movie.find('div', attrs={'class' : 'inline-block ratings-imdb-rating'})
        scores.append(score.text)

        metascore = movie.find('span', attrs={'class' : 'metascore favorable'})
        if metascore is not None:
            metascores.append(metascore.text)
        else:
            metascores.append(-1)

        votes = movie.findAll('span', attrs={'name' : 'nv'})
        votes_numbers.append(votes[0].text)

        if len(votes) > 1:
            gross.append(votes[1].text)
        else:
            gross.append(-1)

    
    
```

```python
base_link = 'https://www.imdb.com'
```

```python
while url != '':
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    movies = soup.findAll('div', attrs={'class' : 'lister-item mode-advanced'})
    extract_movies(movies, titles, ranking, years, durations, votes_numbers, scores, metascores, gross)
    link = soup.find('a', attrs={'class' : 'lister-page-next next-page'})
    if link is not None:
        url = base_link + link['href']
    else:
        url = ''
    print(url)
```

```python
movie_df = pd.DataFrame({'Title': titles, 'Ranking' : ranking, 'Year' : years, 'Time' : durations, 'Votes' : votes_numbers, 'Score' : scores, 'Metascore' : metascores, 'Gross' : gross})
```
